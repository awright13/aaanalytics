In this demo, we will compare 1,000 randomly generated samples.
The elements of each sample will be generated randomly from the set of integers [-5,5], i.e. a uniform distribution over this discrete set of integers.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('alex')

sample_means = []
max_sample_size = 1000


for sample_size in range(1,max_sample_size):
    sample = np.random.randint(low=-5, high=6, size=sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
```

```python
x_index = range(1,max_sample_size)
yticks = [x/2 for x in range(-6,7,1)]

fig = plt.figure(figsize=(14,7))
plt.scatter(x_index,sample_means, marker = 'o', s= 4)
plt.yticks(yticks)
plt.grid(color = 'black', alpha=0.25, axis='y')
plt.title('The Law of Large Numbers \n Convergence of Sample Mean to Population Mean as n-->infinity ', fontsize='20')
plt.ylabel('Sample Mean')
plt.xlabel('Sample Size')
plt.show()
```
