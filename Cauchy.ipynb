```python
import matplotlib.pyplot as plt
from scipy.stats import cauchy
import numpy as np
```

```python
locs = [-5,0,5]
scales = [.5,1,2]


fig,ax  = plt.subplots(figsize=(12,8))

x_space = np.linspace(cauchy.ppf(0.01),
                     cauchy.ppf(0.99), 1000)


for loc in locs:
    for scale in scales:

        ax.plot(x_space, 
                cauchy.pdf(x_space,loc = loc, scale = scale),
                lw=3, alpha=0.6, label = 'cauchy pdf, loc = {}, scale = {}'.format(loc,scale))
plt.title('Your Garden Variety Cauchy pdfs', fontsize= 18)
plt.legend()
plt.show()
```

```python
locs = [-5,0,5]
scales = [.5,1,2]


fig,ax  = plt.subplots(figsize=(12,8))

x_space = np.linspace(cauchy.ppf(0.001),
                     cauchy.ppf(0.999), 10000)


for loc in locs:
    for scale in scales:

        ax.plot(x_space, 
                cauchy.cdf(x_space,loc = loc, scale = scale),
                lw=3, alpha=0.6, label = 'cauchy cdf, loc = {}, scale = {}'.format(loc,scale))
plt.title('Your Garden Variety Cauchy cdfs', fontsize= 18)
plt.legend()
plt.show()
```
