```python
import matplotlib.pyplot as plt
from scipy.stats import cauchy
import numpy as np

plt.style.use('alex')
```

```python
locs = [-5,0,5]
scales = [.5,1,2]


fig,ax  = plt.subplots(figsize=(12,8))

x_space = np.linspace(cauchy.ppf(0.01),
                     cauchy.ppf(0.99), 1000)


for loc in locs:
    for scale in scales:

        ax.plot(x_space, 
                cauchy.pdf(x_space,loc = loc, scale = scale),
                lw=3, alpha=0.6, label = 'cauchy pdf, loc = {}, scale = {}'.format(loc,scale))
plt.title('Your Garden Variety Cauchy pdfs', fontsize= 18)
plt.legend()
plt.show()
```

```python
locs = [-15,0,15]
scales = [.5,1,2]


fig,ax  = plt.subplots(figsize=(12,8))

x_space = np.linspace(cauchy.ppf(0.01),
                     cauchy.ppf(0.99), 10000)


for loc in locs:
    for scale in scales:

        ax.plot(x_space, 
                cauchy.cdf(x_space,loc = loc, scale = scale),
                lw=3, alpha=0.6, label = 'cauchy cdf, loc = {}, scale = {}'.format(loc,scale))
plt.title('Your Garden Variety Cauchy cdfs', fontsize= 18)
plt.legend()
plt.show()
```

Now we will take a look at the distribution of sample means, and see that it in fact does NOT conform the Central Limit Theorem, which states that the distribution of sample means converge to a normal distribution. Instead, the sample means distribute themselves as a cauchy rv with the same parameters as the samples.
As you can see below the distribution of generated sample means fits much closer to the plotted cauchy distribution rather than the standard normal distribution
