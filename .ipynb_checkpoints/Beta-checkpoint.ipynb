```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('alex')
```

```python
a = 150,
b = 219;

mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
print('mean: {:.4f}'.format(mean[0]) + '\n'
      'variance: {:.4f}'.format(var[0]) + '\n'
      'skew: {:.4f}'.format(skew[0]) + '\n'
      'kurtosis: {:.4f}'.format(kurt[0]))
```

First let's look at what happens when we hold alpha constant while varying beta:
The interpretation here would be that we are holding the # number of successes constant, while increasing or decreasing the number of failures.
I would expect then for the peak to shift towards 0, and for variance to decrease as the total number of trials increases.

```python
a_list = np.ones(20)*10
b_list = np.array(range(1,21))
b_list[-1] = 100
x_spaces = [np.linspace(beta.ppf(0.001, a_list[i], b_list[i]), beta.ppf(0.999,a_list[i],b_list[i]),100) for i in range(len(a_list))]
y_spaces = [beta(a_list[i],b_list[i]).pdf(x_spaces[i]) for i in range(len(a_list))]
```

```python
cols = 4;
rows = 5;

fig, ax = plt.subplots(rows,cols,figsize=(16,20))

iterator = 0;

for i in range(rows):
    for j in range(cols):
        ax[i,j].plot(x_spaces[iterator],y_spaces[iterator])
        ax[i,j].set_ylim(0,8)
        ax[i,j].set_xlim(0,1)
        ax[i,j].text(x=0.5 -.02*i,y=6,s='alpha: {:} \nbeta: {:.1f}'.format(a_list[iterator],b_list[iterator]))
        iterator += 1

fig.suptitle('Series of Beta distributions with constant alpha=10, beta=1..19 then 100', fontsize=20, x=0.5, y = 0.9)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15,10))

for i in range(len(x_spaces)):
    plt.plot(x_spaces[i],y_spaces[i])
    plt.title('Beta Distribution with fixed alpha=10, beta=1..19 then 100', fontsize=20)
plt.show()
```

Let's switch now and hold beta constant, while varying alpha.

```python
a_list = np.array(range(1,21))
b_list = np.ones(20)*10
a_list[-1] = 100
x_spaces = [np.linspace(beta.ppf(0.001, a_list[i], b_list[i]), beta.ppf(0.999,a_list[i],b_list[i]),100) for i in range(len(a_list))]
y_spaces = [beta(a_list[i],b_list[i]).pdf(x_spaces[i]) for i in range(len(a_list))]
```

```python
cols = 4;
rows = 5;

fig, ax = plt.subplots(rows,cols,figsize=(15,20))

iterator = 0;

for i in range(rows):
    for j in range(cols):
        ax[i,j].plot(x_spaces[iterator],y_spaces[iterator])
        ax[i,j].set_ylim(0,8)
        ax[i,j].set_xlim(0,1)
        ax[i,j].text(x=0.5 -.02*i,y=6,s='alpha: {:} \nbeta: {:.1f}'.format(a_list[iterator],b_list[iterator]))
        iterator += 1
        
fig.suptitle('Series of Beta distributions with constant beta=10, varying alpha=1..19 then 100', fontsize=20, x=0.5, y = 0.9)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15,10))

for i in range(len(x_spaces)):
    plt.plot(x_spaces[i],y_spaces[i])
    plt.title('Beta Distribution with fixed beta=10, alpha=1..19 then 100', fontsize=20)
plt.show()
```

Now what if we watched the distribution update as we get results from a binomial distribution, i.e. find the Beta distribution after each Bernoulli trial?

```python
p = 0.333;
trials = np.random.binomial(n=1,p=p, size=99)
initial_alpha, a = 6,6;
initial_beta, b= 2,2;
a_list = np.zeros(100);
b_list = np.zeros(100);
a_list[0] = initial_alpha
b_list[0] = initial_beta

for i in range(len(trials)):
    if trials[i] == 0:
        b += 1
    else:
        a +=1
    a_list[i+1] = a
    b_list[i+1] = b

x_spaces = [np.linspace(beta.ppf(0.001, a_list[i], b_list[i]), beta.ppf(0.999,a_list[i],b_list[i]),100) for i in range(len(a_list))]
y_spaces = [beta(a_list[i],b_list[i]).pdf(x_spaces[i]) for i in range(len(a_list))];
```

```python
cols = 5;
rows = 10;

fig, ax = plt.subplots(rows,cols,figsize=(20,40))

iterator = 0;

for i in range(rows):
    for j in range(cols):
        ax[i,j].plot(x_spaces[iterator],y_spaces[iterator])
        ax[i,j].set_ylim(0,8)
        ax[i,j].set_xlim(0,1)
        ax[i,j].text(x=0.5 -.005*i,y=6,s='alpha: {:} \nbeta: {:.1f}'.format(a_list[iterator],b_list[iterator]))
        iterator += 1
        
fig.suptitle('Evolution of Beta Distribution updating from Bernoulli trials with p = 0.333; inital a=6, b=2', fontsize=20, x=0.5, y = 0.9)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15,10))

for i in range(len(x_spaces)):
    plt.plot(x_spaces[i],y_spaces[i], alpha = (i+1)/20)
    plt.title('Evolution of Beta Distribution updating from Bernoulli trials with p = 0.333; inital a=6, b=2', fontsize=20)
plt.show()
```
